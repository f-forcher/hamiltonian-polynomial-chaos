```python
# Harmonic Hamiltonian
def H1(q,p,k0):
    return sy.Number(1)/2*p**2+sy.Number(1)/2*k0*q**2

# Duffing Hamiltonian
def H3(q,p,k0):
    return sy.Number(1)/2*p**2 - sy.Number(1)/2*k0*q**2+sy.Number(1)/4*q**4
```

```python
def ch_to_sy(ch_poly, norm = None):
    sy_tolerance = 10**-10
    num_vars = len(ch_poly.indeterminants)
    num_terms = len(ch_poly.coefficients)
    # Make vars
    sy_vars = []
    #print(ch_poly.names)
    for ch_var_id in [ch_poly.names[i][1:] for i in range(num_vars)]:
        #print(ch_var_id)
        sy.var(f"s{ch_var_id}")
        sy_vars.append(globals()[f"s{ch_var_id}"])
    
    terms = [ch_poly.coefficients[j]*sy.prod([sy_vars[i]**ch_poly.exponents[j,i] 
                          for i in range(num_vars)]) for j in range(num_terms)]
    # nsimplify approximates the float numbers into simpler exact representations
    # https://docs.sympy.org/latest/modules/simplify/simplify.html#nsimplify
    if norm is None:
        poly = sy.nsimplify(sum(terms), full=False, tolerance = sy_tolerance)
    else:
        poly = sy.nsimplify(sum(terms), full=False, tolerance = sy_tolerance) / \
        sy.sqrt(sy.nsimplify(norm, full=False, tolerance = sy_tolerance))
    return poly
```

### Definitions

```python
def gaussdensity(x, mu, sigma, numerical=False):
    if numerical:
        return 0.398942280401433*sy.exp(-0.5*((x-mu)/sigma)**2)
    else:
        return sy.Number(1)/(sigma*sy.sqrt(2*sy.pi))*sy.exp(-sy.Number(1)/2*((x-mu)/sigma)**2 )
```

```python
# Get concrete polynomial expansion
def psic(i, *args):
    #print(args)
    # Should receive as many args as germ variables
    orig_vars = [globals()[f"s{j}"] for j in range(len(args))]
    varsubs = [(orig_vars[j], args[j]) for j in range(len(args))]
    return sy_polys_orig[i].subs(varsubs)
```

```python
### PCE Expansion

#PCE germ variable xi
xi0 = s0
xi1 = s1

PCEvars_list = [Z[i] for Z in (Q,P) for i in range(num_polys)]

#Ham = H3

# Get subset of full PCE expansion, for parallelization
def series_range(z_var, low=0, high=num_polys):
    return sum([z_var[i]*psic(i,xi0,xi1) for i in range(low,high)])


qqcc = series_range(Q, 0, num_polys)#.evalf()
ppcc = series_range(P, 0, num_polys)#.evalf()

qqcc
```

```python
def poisson_bracket(H, g, num_polys=num_polys):
#     Qis = [Q[i] for i in range(num_terms)]
#     Pis = [P[i] for i in range(num_terms)]
    allderiv = sy.Array([sy.derive_by_array(H, Q[i])*sy.derive_by_array(g, P[i]) -
                sy.derive_by_array(H, P[i])*sy.derive_by_array(g, Q[i])
                for i in range(num_polys)])
    return sy.tensorcontraction(allderiv,(0,))
```

```python
def exp_map_matrix(H, g, order, t=t, numerical=False, num_polys=num_polys):
#     Qis = [Q[i] for i in range(num_terms)]
#     Pis = [P[i] for i in range(num_terms)]
    PCEvars_list = [Z[i] for Z in (Q,P) for i in range(num_polys)]
    commutators = [g]
    order = order+1
    for i in range(1,order):
        commutators.append(poisson_bracket(H, commutators[i-1], num_polys=num_polys))
    #exp_map = sy.tensorcontraction(sy.Array([(-t)**i*commutators[i]/sy.factorial(i) for i in range(order)]),(0,)).simplify()
    if numerical:
        exp_map = sy.tensorcontraction(sy.Array([(-t)**i*commutators[i]/sy.factorial(i) for i in range(order)]),(0,))
        out = sy.Array([[exp_map[i][0].evalf() for i in range(2*num_polys)]]).transpose()
    else:
        out = exp_map
#     matrix_map = sy.Matrix([[sy.Poly(exp_map[i][0],PCEvars_list).coeff_monomial(Zi) for Zi in PCEvars_list]
#                         for i in range(2*num_terms)])
    return out
```

```python
identity_g = sy.Array([[Z[i] for Z in (Q,P) for i in range(num_polys)]]).transpose()
identity_g
```
